```python
from sklearn  import linear_model
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:\\charishma\\Andhra_pradesh.csv")
df
```

```python
plt.scatter(df['YEAR'],df['TOTAL IPC CRIMES'])
```

```python
andhra_pradesh_data = df[df['STATE/UT'] == 'ANDHRA PRADESH']
```

```python
andhra_pradesh_data = andhra_pradesh_data[andhra_pradesh_data['YEAR'].between(2001, 2014)]
```

```python
years = andhra_pradesh_data['YEAR'].values.reshape(-1, 1)
total_crimes = andhra_pradesh_data['TOTAL IPC CRIMES']
```

```python
plt.figure(figsize=(10, 5))
plt.plot(years, total_crimes, marker='o', linestyle='-', color='b')
plt.title('Total IPC Crimes in Andhra Pradesh (2001-2014)')
plt.xlabel('Year')
plt.ylabel('Total IPC Crimes')
plt.grid(True)
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(years,total_crimes)
```

```python
reg.predict([[2014]])
```

```python
predictions = reg.predict(years)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(years, total_crimes, color='blue', label='Actual data')
plt.plot(years, predictions, color='red', linestyle='-', linewidth=2, label='Linear regression')
plt.title('Total IPC Crimes in Andhra Pradesh (2001-2014)')
plt.xlabel('Year')
plt.ylabel('Total IPC Crimes')
plt.legend()
plt.grid(True)
plt.show()
```
